```python
# Variables globales
base = dict()
ord = dict()   

# Función para generar conjuntos
def make_set(v):
    base[v] = v
    ord[v] = 0

# Implementación de la función de búsqueda 
# de manera recursiva
def find(v):
    if base[v] != v:
        base[v] = find(base[v])
    return base[v]

# Implementación de la unión de conjuntos
def union(u, v):
    v1 = find(u)
    v2 = find(v)
    if v1 != v2:
        if ord[v1] > ord[v2]:
            base[v2] = v1 
        else:
            base[v1] = v2
            if ord[v1] == ord[v2]: 
                ord[v2] += 1

# Función principal del algoritmo Kruskal
def kruskal(graph):

    # A = {conjunto vacío}
    mst = set()
   
    # Para todo vértice v en G.V
    for v in graph['vertices']:
        make_set(v)
   

    # Ordena la lista G.E en forma no decendente por su peso w
    # En este caso usamos el ordenador dentro de python
    edges = list(graph['edges'])
    edges.sort()
    
    

    # Para toda arista(u,v) en G.E
    for e in edges:
        weight, u, v = e
        # Si encontrar-conjunto(u) != encontrar-conjunto(v)
        if find(u) != find(v):
            # A = A union (u,v)
            union(u, v)
            # Union(u,v)
            mst.add(e)
    return mst 

graph = {
'vertices': ['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O','P','Q'],
'edges': set([
    (7.2,'A','B'),
    (10.42,'A','C'),
    (2.71,'A','D'),
    (6.71,'A','E'),
    (3.17,'A','F'),
    (3.42,'A','G'),
    (7.97,'A','H'),
    (12.13,'A','I'),
    (8.45,'A','J'),
    (7.18,'A','K'),
    (9.43,'A','L'),
    (12.32,'A','M'),
    (11.12,'A','N'),
    (13.79,'A','O'),
    (15.60,'A','P'),
    (15.41,'A','Q'),
    (3.37,'B','C'),
    (4.91,'B','D'),
    (2.33,'B','E'),
    (8.07,'B','F'),
    (6.75,'B','G'),
    (4.53,'B','H'),
    (7.16,'B','I'),
    (6.54,'B','J'),
    (7.42,'B','K'),
    (7.69,'B','L'),
    (8.79,'B','M'),
    (11.76,'B','N'),
    (12.58,'B','O'),
    (13.28,'B','P'),
    (15.69,'B','Q'),
    (7.55,'C','D'),
    (3.10,'C','E'),
    (9.73,'C','F'),
    (7.85,'C','G'),
    (3.67,'C','H'),
    (6.14,'C','I'),
    (5.29,'C','J'),
    (7.32,'C','K'),
    (6.14,'C','L'),
    (6.00,'C','M'),
    (10.11,'C','N'),
    (10.14,'C','O'),
    (9.43,'C','P'),
    (13.14,'C','Q'),
    (4.37,'D','E'),
    (3.20,'D','F'),
    (1.75,'D','G'),
    (4.50,'D','H'),
    (8.27,'D','I'),
    (5.17,'D','J'),
    (4.64,'D','K'),
    (6.03,'D','L'),
    (8.49,'D','M'),
    (8.27,'D','N'),
    (10.39,'D','O'),
    (11.91,'D','P'),
    (12.84,'D','Q'),
    (7.19,'E','F'),
    (5.52,'E','G'),
    (2.06,'E','H'),
    (5.07,'E','I'),
    (4.10,'E','J'),
    (5.47,'E','K'),
    (5.20,'E','L'),
    (6.51,'E','M'),
    (9.21,'E','N'),
    (10.09,'E','O'),
    (10.86,'E','P'),
    (13.89,'E','Q'),
    (1.70,'F','G'),
    (6.78,'F','H'),
    (10.32,'F','I'),
    (6.03,'F','J'),
    (3.60,'F','K'),
    (6.31,'F','L'),
    (7.18,'F','M'),
    (7.60,'F','N'),
    (10.27,'F','O'),
    (11.82,'F','P'),
    (11.05,'F','Q'),
    (4.89,'G','H'),
    (7.70,'G','I'),
    (4.32,'G','J'),
    (2.29,'G','K'),
    (4.55,'G','L'),
    (7.56,'F','M'),
    (5.96,'G','N'),
    (8.57,'G','O'),
    (10.16,'G','P'),
    (10.38,'G','Q'),
    (2.92,'H','I'),
    (1.68,'H','J'),
    (3.79,'H','K'),
    (3.02,'H','L'),
    (4.69,'H','M'),
    (6.98,'H','N'),
    (8.13,'H','O'),
    (8.36,'H','P'),
    (10.81,'H','Q'),
    (3.75,'I','J'),
    (6.06,'I','K'),
    (4.05,'I','L'),
    (2.91,'I','M'),
    (7.09,'I','N'),
    (6.88,'I','O'),
    (6.31,'I','P'),
    (9.98,'I','Q'),
    (2.48,'J','K'),
    (3.66,'J','M'),
    (4.91,'J','N'),
    (6.11,'J','O'),
    (6.95,'J','P'),
    (8.85,'J','Q'),
    (2.70,'K','L'),
    (5.29,'K','M'),
    (4.02,'K','N'),
    (6.19,'K','O'),
    (7.63,'K','P'),
    (8.06,'K','Q'),
    (3.01,'L','M'),
    (4.16,'L','N'),
    (4.89,'L','O'),
    (5.52,'L','P'),
    (7.69,'L','Q'),
    (4.70,'M','N'),
    (3.99,'M','O'),
    (3.34,'M','P'),
    (2.71,'M','Q'),
    (3.07,'N','O'),
    (5.54,'N','P'),
    (4.34,'N','Q'),
    (2.78,'O','P'),
    (3.15,'O','Q'),
    (4.77,'P','Q'),













  

  
  




    ])
}
        


k = kruskal(graph)
print ("Resultado del Minimum Spanning Tree es :"+ str(k)+"\n")
```
